```python
import getpass
import re
import random
import string


def get_password(length = 12):
    """
    Generate a strong password of a specified length.
    
    Args:
        length(int): the length of the password. Defaults to 12
    Returns:
        str: A strong password
    """
    punctutation = "!\|#$%&^()*+,-.?/:;<=>@[|\]^_{}"
#     variable that contain all the special character
    character = string.ascii_letters + string.digits + punctutation
#     join all the letters, the digit and punctation which is been imported from the string module

    return "".join(random.choice(character) for _ in range(length) )
    

     
errors = []   
    
def check_password_strength(password):
    """
    Checks the strength of a password.
    
    Args:
        password(str): The password to check
    Returns:
        list: A list of errors if the password is weak, otherwise an empty list
    """
# check if the password has up to character, has a captical letter, small letter and a digit
    if len(password) < 8:   
        errors.append("Password should be at least eight characters long.")
    if not re.search("[a-z]",password):
        errors.append("Password should contain at least one lowercase letter.")
    if not re.search("[A-Z]",password):
        errors.append("Password should contain at least one uppercase letter.")
    if not re.search("[0-9]",password):
        errors.append("Password must contain a digit")
    if not re.search("[A-Za-z0-9]",password):
        errors.append("Password should contain a special character.")
    return errors

password_running = True       
while password_running:
    
#     get the user_password

    password = getpass.getpass("Please enter your password: ")
    check_password_strength(password)
    
# check if there is no error

    if (errors == []):
        print("Password is strong")
        break
    else:
        for error in errors:
#             shows the error
            print(error)
#     prompt the user , if they want a suggested password
            sugguest = input("Would you like a password suggestion?(y/n) ")
            if sugguest.lower() == "y":
                print(f"Suggested strong password {get_password(12)}")
                password_running = False

        
        
        
        
        
        
        
        
        
        
```
